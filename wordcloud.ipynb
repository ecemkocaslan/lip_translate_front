```python
from wordcloud import WordCloud, ImageColorGenerator
import numpy as np
from PIL import Image

# Read the lips image
lips_mask = np.array(Image.open("/Users/ecemkocaslan/Downloads/redd_lipss-removebg-preview.png"))
#lips_mask = np.array(Image.open("lips_shape.png"))

wordcloud = WordCloud(background_color="white", mask=lips_mask, contour_width=1, contour_color='red').generate(text)

image_colors = ImageColorGenerator(lips_mask)

# Display the generated word cloud
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.imshow(wordcloud.recolor(color_func =image_colors), interpolation='bilinear')
plt.axis("off")
plt.savefig("lips.png")
plt.show()
```
